```python
import pandas as pd


def load_data(path: str) -> pd.DataFrame:
    return pd.read_csv(path)


def basic_report(df: pd.DataFrame) -> dict:
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame")

    if df.empty:
        raise ValueError("DataFrame cannot be empty")

    return {
        'rows': len(df),
        'columns': len(df.columns),
        'column_names': list(df.columns),
        'data_types': df.dtypes.astype(str).to_dict(),
        'missing_values': int(df.isna().sum().sum()),
        'duplicates': int(df.duplicated().sum()),
        'target_distribution': (
            df['target'].value_counts().to_dict()
            if 'target' in df.columns else {}
        ),
        'numeric_summary': df.describe().to_dict()
    }


if __name__ == '__main__':
    path = 'data-2.csv'   # lokalny plik w tym samym folderze
    data = load_data(path)
    print(basic_report(data))
```

Skrypt w podstawowej wersji pokazuje ogólne informacje o danych: liczbę wierszy, kolumn, braków danych i rozkład kolumny target. AI zaproponowało rozbudowanie raportu o typy kolumn, nazwy kolumn, duplikaty i statystyki liczbowe. Dodałem własną walidację, która sprawdza, czy dane są obiektem DataFrame i czy tabela nie jest pusta. Dzięki temu kod jest bezpieczniejszy i łatwiej wykryć błędne dane wejściowe.
