```python
def average_age(records):
    if not records:
        raise ValueError("Lista nie może być pusta")

    return sum(record['age'] for record in records) / len(records)


if __name__ == '__main__':
    sample = [
        {'name': 'Anna', 'age': 21},
        {'name': 'Piotr', 'age': 23},
        {'name': 'Maria', 'age': 25},
    ]

    print(average_age(sample))  # 23.0
```
