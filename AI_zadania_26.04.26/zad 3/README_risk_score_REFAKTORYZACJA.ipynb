# Refaktoryzacja pliku risk_score.py

W ramach zadania wybrałem plik `risk_score.py`.

Funkcja `calculate_risk_score` oblicza uproszczony wynik ryzyka na podstawie:
- wieku,
- poziomu cholesterolu,
- maksymalnego tętna.

Poprosiłem AI o refaktoryzację kodu. AI zaproponowało by ulatwic kod poprzez jego rozbicie w celu latwiejszych komunikatow bledow, oraz kazdy warunek "if" osobno.

Po zmianach kod jest bardziej czytelny, ponieważ łatwiej zobaczyć, która część wzoru odpowiada za wiek, cholesterol i tętno. Dodatkowo komunikaty błędów są bardziej precyzyjne.

Sprawdziłem propozycję AI i poprawiłem opis własnymi słowami. Refaktoryzacja nie zmienia działania funkcji, tylko poprawia czytelność kodu.
