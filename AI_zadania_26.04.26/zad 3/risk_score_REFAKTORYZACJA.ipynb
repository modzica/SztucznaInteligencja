```python
def calculate_risk_score(age: int, cholesterol: int, max_heart_rate: int) -> float:
    if age < 0:
        raise ValueError("Age must be positive.")
    if cholesterol < 0:
        raise ValueError("Cholesterol must be positive.")
    if max_heart_rate <= 0:
        raise ValueError("Max heart rate must be greater than zero.")

    age_score = age * 0.2
    cholesterol_score = cholesterol * 0.05
    heart_rate_score = max_heart_rate * 0.03

    score = age_score + cholesterol_score - heart_rate_score

    return round(score, 2)
    
```
